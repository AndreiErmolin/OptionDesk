```python
import pandas as pd
d = pd.DataFrame({
    'x': [6.1, 9.2, 11.5, 8.6, 12.1, 3.9, 8.4, 10.1, 9.4, 8.9],
    'y': [7.3, 8.7, 12.6, 6.8, 12, 6.9, 10.2, 17.1, 9.2, 11.6]
})
```

```python
from scipy.stats import norm
from scipy.stats import t
import math
def z_param_diff(x1,x2,sigma1,sigma2,n1,n2):
    return (x1-x2)/(math.sqrt(sigma1*sigma1/n1+sigma2*sigma2/n2))
z=z_param_diff(d['x'].mean(),d['y'].mean(),d['x'].std(),d['y'].std(),10,10)
p=1-(t.cdf(-z,17)-t.cdf(z,17))
display(z)
display(p)

```

```python
from scipy.stats import norm
from scipy.stats import t
import math
def z_param_p(p1,n1,p2,n2):
    '''Функция для рассчета разности парметров пропорции
    1.p1-отслеживаемое количсетво в первой вборке
    2.n1-размер первой выборки
    3.p2-отслеживаемое количество во второй выборке
    4.т2-размер второй выборки'''
    p=(p1+p2)/(n1+n2)
    return (p1/n1-p2/n2)/(math.sqrt(p*(1-p))*math.sqrt(1/n1+1/n2))
z=z_param_p(62,100,29,100)
p=1-norm.cdf(z)
display(z)
display(p)
```
